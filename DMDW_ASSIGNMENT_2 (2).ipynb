```python
import pandas as pd
url="https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names=['sepal-length','sepal-width','petal-length','petal-width','class']
dataset =pd.read_csv(url,names=names)
```

```python
print(dataset.shape)
```

```python
print(dataset.head(30))
```

```python
import numpy as np
irisflower = []
for i in range(0, dataset.shape[0]):
    for j in range(0, dataset.shape[1]):
        irisflower.append(dataset.values[i,j])
irisflower = np.array(irisflower)
print(irisflower)
```

```python

df = pd.DataFrame(irisflower, columns=["sepal-width"]) 
df["incident_count"] = 2
indexNames = df[df['sepal-width'] == "nan" ].index
df.drop(indexNames , inplace=True)
df_table = df.groupby("sepal-width").sum().sort_values("incident_count", ascending=False).reset_index()
df_table.head(5).style.background_gradient(cmap='Blues')
```

```python
irisflower = []
for i in range(dataset.shape[0]):
    irisflower.append([str(dataset.values[i,j]) for j in range(dataset.shape[1])])
irisflower = np.array(irisflower)
from mlxtend.preprocessing import irisflowerEncoder
te = irisflowerEncoder()
te_ary = te.fit(irisflower).transform(irisflower)
dataset = pd.DataFrame(te_ary, columns=te.columns_)

dataset.head()
```

```python
first20 = df_table["sepal-width"].head(20).values 
dataset = dataset.loc[:,first20] 
dataset.shape
```

```python
from mlxtend.frequent_patterns import fpgrowth
res=fpgrowth(dataset,min_support=0.05, use_colnames=True)
res.head(10)
```

```python
from mlxtend.frequent_patterns import association_rules
res=association_rules(res, metric="lift", min_threshold=1)
res
```

```python
res.sort_values("confidence",ascending=False)
```
